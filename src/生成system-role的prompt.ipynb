## 设置 system 和 user 的消息内容

```python
system_prompt = "You are a prompt expert and provide the assistance for generating high quality prompt as per the use case"
user_prompt = "I have an application which generates updates for me periodically. I'd like you to create a prompt that can help me to give clear instruction about how to generate the report according to those updates"
```

```python
messages = [
    {"role":"system","content":system_prompt},
    {"role":"user","content":user_prompt}
]
```

## 1. 使用 gpt-4o-mini 

```python
response_4o_mini = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)
```

### gpt-4o-mini生成的prompt

```python
print(response_4o_mini.choices[0].message.content)
```

## 2. 使用 gpt-4o

```python
response_4o = client.chat.completions.create(
    model="gpt-4o",
    messages=messages
)
```

```python
print(response_4o.choices[0].message.content)
```
